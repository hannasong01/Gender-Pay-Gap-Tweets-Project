```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import matplotlib.pyplot as plt

geometry = [Point(xy) for xy in zip(df['longitude'], df['latitude'])]
gdf = GeoDataFrame(df, geometry=geometry)   

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'), cmap='OrRd')
gdf.plot(ax=world.plot(figsize=(15, 15)), marker='o', color='pink', markersize=15);
```

```python
import plotly.express as px
import pandas as pd

fig = px.scatter_geo(df,lat='latitude',lon='longitude', hover_name="location")
fig.update_layout(title = 'Distribution of Tweets in different areas', title_x=1)
fig.show()
```

```python
fig.write_html("world_map.html")
```
